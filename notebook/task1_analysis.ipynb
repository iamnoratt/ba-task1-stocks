```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8-whitegrid')
```

```python
jushi = pd.read_csv('../data/china-jushi_600176.csv', parse_dates=['date'])
zhongji = pd.read_csv('../data/zhongji-xuchuang_300308.csv', parse_dates=['date'])

jushi.head(), zhongji.head()
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(jushi['date'], jushi['close'], label='China Jushi')
ax.plot(zhongji['date'], zhongji['close'], label='Zhongji Xuchuang')
ax.set_title('Close Price Comparison')
ax.set_xlabel('Date')
ax.set_ylabel('Close Price')
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
jushi_norm = jushi['close'] / jushi['close'].iloc[0] * 100
zhongji_norm = zhongji['close'] / zhongji['close'].iloc[0] * 100
ax.plot(jushi['date'], jushi_norm, label='China Jushi (Base=100)')
ax.plot(zhongji['date'], zhongji_norm, label='Zhongji Xuchuang (Base=100)')
ax.set_title('Normalized Performance')
ax.legend()
plt.show()
```
